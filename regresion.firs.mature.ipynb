```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# data 
# y = 2*1 - 1
x=np.array([1,2,3,4]).reshape(-1,1)
y=np.array([1,3,5,7])
```

```python
 #model 
from sklearn.linear_model import LinearRegression 
model = LinearRegression ()
model.fit(x,y)
```

```python
m = model.coef_ [0] #number of x 
c = model.intercept_  #arz az mabda

print(m ,c )
```

```python
# plot

plt.plot(x,y , '*')
plt.plot(x, model.predict(x))
plt.show()
 
```

```python
result=model.predict([[5]])
result
```
